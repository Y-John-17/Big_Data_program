```python
from datascience import *
import matplotlib.pyplot as plt
import numpy as np

#import matplotlib
#matplotlib.use('TkAgg')
%matplotlib inline

rank1 = {}
rank2 = {}
rank3 = {}
rank4 = {}
rank5 = {}
rank6 = {}
rank7 = {}

def extract_rank(data,name):
    temp = data.where('姓名',are.equal_to(name)).column(11).item(0)
    return temp
    

data1 = Table.read_table('./test/20211105-utf-8.csv')
data2 = Table.read_table('./test/20211106-utf-8.csv')
data3 = Table.read_table('./test/20211107-utf-8.csv')
data4 = Table.read_table('./test/20211108-utf-8.csv')
data5 = Table.read_table('./test/20211109-utf-8.csv')
data6 = Table.read_table('./test/20211109-1-utf-8.csv')
data7 = Table.read_table('./test/20211110-utf-8.csv')

name1 = '周靖珊'
name2 = '黄思哲'


rank1[0] = extract_rank(data1,name1)
rank2[0] = extract_rank(data2,name1)
rank3[0] = extract_rank(data3,name1)
rank4[0] = extract_rank(data4,name1)
rank5[0] = extract_rank(data5,name1)
rank6[0] = extract_rank(data6,name1)
rank7[0] = extract_rank(data7,name1)

rank1[1] = extract_rank(data1,name2)
rank2[1] = extract_rank(data2,name2)
rank3[1] = extract_rank(data3,name2)
rank4[1] = extract_rank(data4,name2)
rank5[1] = extract_rank(data5,name2)
rank6[1] = extract_rank(data6,name2)
rank7[1] = extract_rank(data7,name2)

rank_array1 = make_array(rank1[0],rank2[0],rank3[0],rank4[0],rank5[0],rank6[0],rank7[0])
rank_array2 = make_array(rank1[1],rank2[1],rank3[1],rank4[1],rank5[1],rank6[1],rank7[1])

results = Table().with_columns('考试序号',make_array(1,2,3,4,5,6,7),'姓名',rank_array1)
results.show()
results = Table().with_columns('考试序号',make_array(1,2,3,4,5,6,7),'姓名',rank_array2)
results.show()

plt.plot(make_array(1,2,3,4,5,6,7),rank_array1,'o--',label=name1)
plt.plot(make_array(1,2,3,4,5,6,7),rank_array2,'o--',label=name2)

plt.xlabel('历次考试')
plt.ylabel('年级RANK')

plt.axis([0,10,650,150])
plt.legend(loc='upper right')

plt.rcParams['font.sans-serif']=['Microsoft YaHei']

plt.grid()
plt.show()
```
